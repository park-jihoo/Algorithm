```python
import os
import regex

# get the current working directory
os.getcwd()

# ls
os.chdir('/Users/Jihoo/Algorithm/leetcode')
readmes = []
# get all readme files in the current working directory and subdirectories
for dirpath, dirnames, filenames in os.walk(os.getcwd()):
    if 'README.md' in filenames and dirpath.split('/')[-2] == 'leetcode':
        if dirpath.split('/')[-1][:3].isdigit() and dirpath.split('/')[-1][4] == '-':
            readmes.append(os.path.join(dirpath, 'README.md')) # path
```

```python
import markdown

# read the readme files
readme_contents = []
for readme in readmes:
    with open(readme, 'r') as f:
        readme_contents.append((f.read(), readme))
        
# convert to html
html_contents = []
for readme_content, filepath in readme_contents:
    html_contents.append((markdown.markdown(readme_content), filepath))
    
from bs4 import BeautifulSoup

# parse html
soup_contents = []
for html_content, filepath in html_contents:
    soup_contents.append((BeautifulSoup(html_content, 'html.parser'), filepath))
    print(filepath)
```

```python
for soup_content, filepath in soup_contents:
    try:
        difficulty = soup_content.find('img')['alt'][12:]
        splitedpath = filepath.split('/')[:-1]
        oldpath = '/'.join(splitedpath)
        splitedpath.insert(-1, difficulty)
        newpath = '/'.join(splitedpath)
        os.mkdir(newpath)
        os.rename(oldpath, newpath)
    except AttributeError:
        pass
    except TypeError:
        pass    
    except KeyError:
        pass
    except FileNotFoundError:
        pass
    except FileExistsError:
        print(filepath)
```
